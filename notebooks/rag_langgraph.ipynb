```python
!pip install langchain-community
import os
import sys

from IPython.display import Markdown, display
from langchain_community.vectorstores import Chroma

current_dir = os.getcwd()
kit_dir = os.path.abspath(os.path.join(current_dir, '..'))
repo_dir = os.path.abspath(os.path.join(kit_dir, '..'))

sys.path.append(kit_dir)
sys.path.append(repo_dir)

from src.document_retrieval import DocumentRetrieval
from src.langgraph_rag import RAG

CONFIG_PATH = os.path.join(kit_dir, 'config.yaml')
```

```python
documentRetrieval = DocumentRetrieval()
*_, embedding_model_info, retrieval_info, _ = documentRetrieval.get_config_info()
embeddings = documentRetrieval.load_embedding_model()
```

```python
vectorstore = Chroma(persist_directory=os.path.join(kit_dir, 'data/my_vector-db'), embedding_function=embeddings)
```

```python
rag = RAG(config_path=CONFIG_PATH, embeddings=embeddings, vectorstore=vectorstore)

# Initialize chains
rag.init_llm()
rag.init_qa_chain()
rag.init_final_generation()

# Build nodes
workflow = rag.create_rag_nodes()
print(workflow)

# Build graph
rag.build_rag_graph(workflow)
```

```python
rag.display_graph(rag.app)
```

```python
response = rag.call_rag(question='user question')

display(Markdown(response))
```
